```python
import requests
import csv

# Definir o termo de pesquisa
search_term = []
limit = 50
total_results = 150
ids = []
TitleGeneric=[]
all_item_data=[]

# Calcular o número de chamadas necessárias (neste caso, 3 chamadas)
pages = total_results // limit
```

#Pesquisa no grupo de televisores para encontrar outros 3 ID



search_generic = f"https://api.mercadolibre.com/sites/MLA/search?q=televisores&limit={limit}"


```python
# Descobrir outros ID para comparação
for page in range(pages):
    url = f"https://api.mercadolibre.com/sites/MLA/search?q=televisores&limit={limit}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        items = data.get('results', [])
        
        for item in items:
            TitleGeneric.append(item['title'])  # Adiciona o nome do item à lista
        
    else:
        print(f"Erro ao acessar a página {page + 1}")

```

```python
# Mostrar os Nomes coletados
TitleGeneric
```

```python
#Nomes escolhidos
search_terms=['Smart Tv Quint 32 Pulgadas Qt3-32gtv2024 Google Tv Lh', 'Smart Tv Kanji Kj-32mt005 Led Hd 32 Android Tv Hey Google', 'Smart Tv Tcl 40 S5400 Full Hd Android Tv Hdr 10']
```

```python
# Percorrer os nomes selecionados e as paginas
for search_term in search_terms:
    for page in range(pages):
            url = f"https://api.mercadolibre.com/sites/MLA/search?q={search_terms}&limit={limit}&offset={page * limit}"
            response = requests.get(url)
            
            if response.status_code == 200:
                data = response.json()
                items = data.get('results', [])
                
                for item in items:
                    ids.append(item['id'])  # Adiciona o ID do item à lista
                
            else:
                print(f"Erro ao acessar a página {page + 1}")
    
```

```python

# Lista de IDs dos itens
item_ids = ids
```

```python
item_ids
```

```python

# Requisição GET para obter mais detalhes em cada ID salvo
for item_id in item_ids:
    url = f"https://api.mercadolibre.com/items/{item_id}"
    
    # Realiza a requisição GET
    response = requests.get(url)
    
    if response.status_code == 200:
        # Se a resposta for bem-sucedida, processa os dados
        item_data = response.json()
        # Criando dicionario para colocar no csv
        item_info = {
            'id': item_data.get('id'),
            'title': item_data.get('title'),
            'price': item_data.get('price'),
            'currency_id': item_data.get('currency_id'),
            'available_quantity': item_data.get('available_quantity'),
            'condition': item_data.get('condition'),
            'seller_id': item_data.get('seller', {}).get('id'),  # 'seller' é um dicionário
            'category_id': item_data.get('category_id')
        }

        all_item_data.append(item_info)

        print(all_item_data)  
    else:
        print(f"Erro ao obter dados para o Item_ID {item_id}: {response.status_code}")
```

```python
all_item_data
```

```python
csv_file = 'API_Mercado_Livre.csv'

fields = item_info

# Abrir o arquivo CSV para escrita
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fields)
    
    # Escrever o cabeçalho (nomes das colunas)
    writer.writeheader()
    
    # Escrever os dados de todos os itens
    for item in all_item_data:
        writer.writerow({field: item[field] for field in fields})

print(f"Dados salvos com sucesso no arquivo {csv_file}")
```
